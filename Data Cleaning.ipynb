```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/OmriBisa/Python-Pandas-Data-Cleaning-Project/main/Customer_List.csv', encoding='latin-1')
df
```

```python
dfVIP = pd.read_csv('https://raw.githubusercontent.com/OmriBisa/Python-Pandas-Data-Cleaning-Project/main/VIP_Customer.csv')
dfVIP
```

```python
#Let's merge the two datasets
newdf = df.merge(dfVIP, how='left', on='Customer_ID')
newdf
```

```python
#Delete duplicate values
newdf=newdf.drop_duplicates(subset='Customer_ID')

#Drop columns that will not be used by the sales team
newdf=newdf.drop(columns = "Last_Purchase_ID")

#There are several unnecessary characters in the Last_Name field both before and after the name.
#The following code will be used to clear the names:
newdf["Last_Name"] =newdf["Last_Name"].str.strip("123._/")

newdf
```

```python
#There is no standard format for the Phone_Number column.
#The sales team has asked that the numbers be arranged in a uniform format.

#Remove unwanted characters.
newdf["Phone_Number"] = newdf["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
 

# nan / Na replace
newdf["Phone_Number"] = newdf["Phone_Number"].str.replace('Na','')
newdf["Phone_Number"] = newdf["Phone_Number"].fillna(" ")

#Creating a uniform format for the Phone_Number column.
newdf["Phone_Number"] = newdf["Phone_Number"].apply(lambda x: str(x))
newdf["Phone_Number"] = newdf["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
newdf["Phone_Number"] = newdf["Phone_Number"].str.replace('--','')

newdf
```

```python
#We need to divide the Address column into three columns:  "Zip_Code" ,"Street" and "State" 
#Since the sales team wants to approach clients based on certain states.
#The data can then be filtered in accordance with the team's requests.

newdf[[ 'Zip_Code', 'State','Street']]=newdf['Address'].str.split(pat=',', expand=True)
newdf['State'] = newdf['State'].str.strip()

newdf
```

```python
#Only Customers who meet the following requirements will be contacted by the sales team: 
#yes = VIP_Customer; and in addition , no = Request_To_Unsubscribe.
#Using the following code, will create consistent values in order to filter 'N'/'Y'.

newdf["VIP_Customer"]=newdf["VIP_Customer"].str.replace('No','N')
newdf["VIP_Customer"]=newdf["VIP_Customer"].str.replace('Yes','Y')

newdf["Request_To_Unsubscribe"]=newdf["Request_To_Unsubscribe"].str.replace('No','N')
newdf["Request_To_Unsubscribe"]=newdf["Request_To_Unsubscribe"].str.replace('Yes','Y')

# Change cells with NaN to empty cell ('')
newdf=newdf.fillna('')

#Drop customers who are: Yes=Request_To_Unsubscribe
#Display those who answered NO or the cell is empty (ie did not answer)
for x in newdf.index:
    if newdf.loc[x, "Request_To_Unsubscribe"] == 'Y':
        newdf.drop(x, inplace=True)
        
# Drop customers without a number in the Phone_Number column.
newdf = newdf[newdf['Phone_Number'].str.strip() != '']

for x in newdf.index:
    if newdf.loc[x, "Phone_Number"]== '':
        newdf.drop(x, inplace=True)

# Filter only customers on the condition: Yes = VIP_Customer
newdf=newdf[newdf['VIP_Customer']=='Y']

newdf
```

```python
#I'm using the code below to display all of the rows and columns.
pd.set_option('display.max_row', None)
pd.set_option('display.max_columns', None)

#Filter and display only customers from the following states: New York, Pennsylvania,Texas.
# Sort First_Name in descending order.

Specific_State = ['New York','Pennsylvania','Texas']

filt=newdf['State'].isin(Specific_State)

newdf.loc[filt].sort_values(by='First_Name', ascending=False)
```
